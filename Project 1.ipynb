```python
f = open('words.txt','w')
d = {}
json.dump(d,f)
f.close
```

```python
while True:
    Item = int(input('Enter the Choice :'))
    if Item==1:
        f = open('words.txt','w')
        word = input('Enter the word :')
        meaning = input('Enter the meaning :')
        d[word] = meaning
        json.dump(d,f)
        f.close
    elif Item==2:
        f = open('words.txt','r')
        find = json.load(f)
        find_meaning = input('Enter the word :')
        if find_meaning in d:
            print(d[find_meaning])
        else:
            print('Sorry,The word is not found!!')
        json.dump(d,f)
        f.close()
    elif Item==3:
        f = open('words.txt','w')
        word1 = input('Enter a word :')
        update = input('Enter the new meaning :')
        d[word1]=update
        json.dump(d,f)
        f.close()
    elif Item==4:
        print('Thank you for visiting!!')
    elif Item >4 or Item < 4:
        print('Please select Number 1 to 4')
    break
```

# 1B Library Book Management System

```python
#-----------------------------------------Library----------------------------------------------------------------------------------
import sqlite3
from tkinter import *
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessageBox

#--------------------------------------------------------------------------------------------------------------------------------
root = Tk()
root.title('Library Book Management System')

#---------------------------------------Variable-----------------------------------------------------------------------------------------

BOOKTITLE = StringVar()
AUTHOR = StringVar()
YEAR = StringVar()
ISBN = StringVar()

#----------------------------------------Function----------------------------------------------------------------------------------------------

def Database():
    global conn,cursor
    conn = sqlite3.connect('lbm.db')
    cursor = conn.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS lbm(booktitle TEXT,author TEXT,year TEXT,isbn TEXT)")

def Add():
    Database()
    if BOOKTITLE.get()=="" or AUTHOR.get()=="" or YEAR.get()=="" and ISBN.get()=="":
        result.config(text='Please Enter All Data',fg='red')
    else:
        cursor.execute("INSERT INTO lbm values(?,?,?,?)",(BOOKTITLE.get(),AUTHOR.get(),YEAR.get(),ISBN.get()))
        conn.commit()
        BOOKTITLE.set("")
        AUTHOR.set("")
        YEAR.set("")
        ISBN.set("")
        cursor.close()
        conn.close()
        result.config(text='Record Inserted',fg='green')

def Display():
    tree.delete(*tree.get_children())
    Database()
    cursor.execute('SELECT * FROM lbm ')
    row = cursor.fetchall()
    for data in row:
        tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
    conn.commit()
    cursor.close()

def Search():
    tree.delete(*tree.get_children())
    Database()
    cursor.execute("SELECT * FROM lbm WHERE booktitle = ? or author = ? or year = ? or isbn = ?",[BOOKTITLE.get(),AUTHOR.get(),YEAR.get(),ISBN.get()])
    rows=cursor.fetchall()
    for data in rows:
        tree.insert('','end',values=(data[0],data[1],data[2],data[3]))
    conn.commit()
    cursor.close()

def Issue():
    Database()
    result.config(text='Book is Issue.Thank you',fg='black')

def Delete():
    Database()
    if BOOKTITLE.get()=='' and ISBN.get()=='':
        result.config(text='Please Enter Booktitle or Isbn to Delete.',fg='red') 
    else:
        cursor.execute("DELETE FROM lbm  WHERE  BOOKTITLE = ? or AUTHOR=? or YEAR=? or ISBN=?",[BOOKTITLE.get(),AUTHOR.get(),YEAR.get(),ISBN.get()])
        conn.commit()  
        cursor.close()
        result.config(text='Successfully Deleted !!!',fg='green')
    

def Exit():
    exit = tkMessageBox.askquestion('Exit',message='Do you want to Exit?',icon='warning') 
    if exit == 'yes':
        root.destroy()
        
#--------------------------------------Frames----------------------------------------------------------------------------------------------

Top = Frame(root,width=800,height=80,bd=8,bg='red',relief='groove')
Top.pack(side = TOP)

Left = Frame(root,width=400,height=600,bd=8,bg='pink',relief='raise')
Left.pack(side = LEFT)

Right = Frame(root,width=600,height=600,bd=8,relief='raise')
Right.pack(side = RIGHT)

Data = Frame(Left,width=300,height=600,bd=8)
Data.pack(side = TOP)

But = Frame(Left,width=200,height=100,relief='raise')
But.pack(side = BOTTOM)

#------------------------------------Label-----------------------------------------------------------------------------------------------------------

title_txt = Label(Top,width=900,font=('arial',24),text='Library Management Application')
title_txt.pack()

booktitle_txt = Label(Data,font=('arial',15),text='Book Title :',bd=15)
booktitle_txt.grid(row=0)

author_txt = Label(Data,font=('arial',15),text='Author :',bd=15)
author_txt.grid(row=1)

year_title = Label(Data,font=('arial',15),text='Year :',bd=15)
year_title.grid(row=2)

isbn_txt = Label(Data,font=('arial',15),text='ISBN :',bd=15)
isbn_txt.grid(row=3)

result = Label(But)
result.pack(side=TOP)

#------------------------------------Entry-------------------------------------------------------------------------------------------------------------------------
Book_title = Entry(Data,textvariable=BOOKTITLE,width=30)
Book_title.grid(row=0,column=1)

author = Entry(Data,textvariable=AUTHOR,width=30)
author.grid(row=1,column=1)

year = Entry(Data,textvariable=YEAR,width=30)
year.grid(row=2,column=1)

isbn = Entry(Data,textvariable=ISBN,width=30)
isbn.grid(row=3,column=1)

#-----------------------------------Buttons---------------------------------------------------------------------------------------------------------------------------------

but_read = Button(But,width=10,text='Display',command=Display)
but_read.pack(side=LEFT)

but_search = Button(But,width=10,text='Search',command=Search)
but_search.pack(side=LEFT)

but_create = Button(But,width=10,text='Add',command=Add)
but_create.pack(side=LEFT)

but_issue = Button(But,width=10,text='Issue',command=Issue)
but_issue.pack(side=LEFT)

but_delete = Button(But,width=10,text='Delete',command=Delete)
but_delete.pack(side=LEFT)

but_exit = Button(But,width=10,text='Exit',command=Exit)
but_exit.pack(side=LEFT)

#----------------------------------------------------------------------------------------------------------------------------------------------------------
tree = ttk.Treeview(Right,columns=('BookTitle','Author','Year','Isbn'))


Scrollbary = Scrollbar(Right,orient=VERTICAL)
Scrollbary.config(command=tree.yview)
Scrollbary.pack(side=RIGHT,fill=Y)


tree.heading('BookTitle',text='Booktitle',anchor=W)
tree.heading('Author',text='Author',anchor=W)
tree.heading('Year',text='Year',anchor=W)
tree.heading('Isbn',text='Isbn',anchor=W)


tree.column('#0',stretch=NO,minwidth=50,width=20)
tree.column('#1',stretch=NO,minwidth=50,width=120)
tree.column('#2',stretch=NO,minwidth=50,width=100)
tree.column('#3',stretch=NO,minwidth=50,width=200)
tree.column('#4',stretch=NO,minwidth=50,width=200)

tree.pack()
root.mainloop()
```
